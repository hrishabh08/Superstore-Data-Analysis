#### Load Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Load the Data

```python
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

#### Drop Unwanted columns

```python
df.drop(columns=['Postal Code'] , inplace=True)
df.head()
```

#### Understanding the dataset

```python
# Checking uniques for differnt columns to understand the data
print("Different shipping mode for items:",df['Ship Mode'].unique())
print("Differnt segments for stores :",df['Segment'].unique())
print("Countries:",df['Category'].unique())
print("Different category of items:",df['Category'].unique())
print("Different sub-category of items:",df['Sub-Category'].unique())
print("Differnt regions:",df['Region'].unique())
```

```python
# Statistical description of data
df.describe()
```

```python
# More information about data
df.info()
```

```python
# Check if aay missing values in dataset
df.isna().sum()
```

So, No missing values in the dataset

### Analysis of the data

```python
# Compare the sales based on region
df.groupby('Region')['Sales'].sum().plot.bar()
plt.xlabel('Regions')
plt.ylabel("Sales")
plt.title("Sales based on Regions")
plt.show()
```

```python
# Understanding profit
df.groupby('Region')['Profit'].sum().plot.bar()
plt.xlabel('Regions')
plt.ylabel("Profit")
plt.title("Profit based on Regions")
plt.show()
```

```python
# Understanding market share of sales and profit for each region
df.groupby('Region')['Sales'].sum().plot.pie(autopct="%0.2f")
plt.legend()
```

```python
df.groupby('Region')['Profit'].sum().plot.pie(autopct="%0.2f")
plt.legend()
```

```python
#Analysis of Sales based on segment
df.groupby('Segment')['Sales'].sum().plot.bar()
plt.xlabel('Segments')
plt.ylabel("Sales")
plt.title("Sales based on Segments")
plt.show()
```

```python
#Analysis of Profit based on segment
df.groupby('Segment')['Profit'].sum().plot.bar()
plt.xlabel('Segments')
plt.ylabel("Profit")
plt.title("Profit based on Segments")
plt.show()
```
